### first visualization

```python
#plotting data of 'Year acquired' and 'Year constructed'

buildings.plot(x = 'Year Acquired', y='Year Constructed', color='purple', figsize=(20, 6), kind='scatter')
```

The relationship between 'Year Acquired' and 'Year Constructed' is positive except some outliers.

### using groupby to observe their relationship

```python
for grouped in buildings.groupby("Year Acquired"):
    print(grouped[0], grouped[1]["Year Constructed"])
    #print(grouped)
    print('----------------------------')
#Here I tried to groupby 'Year Acquired' and find 'Year Constructed' by using duple location method.
#However, because 'Year Acquired' is number not string, it is hard to tell which is squence or years.
#But still can find out which is 'Year Acquired' and 'Year Constructed'
#The left one is 'Year Acquired', the middle one is sequence, and the right one is 'Year Constructed'
```

```python
print(buildings["Year Acquired"].min(), buildings["Year Acquired"].max())
print(buildings["Year Constructed"].min(), buildings["Year Constructed"].max())
#The 1753 year is min and 2019 year is max for both of them
```

```python
aggre_acquire=buildings.groupby("Year Acquired")["Year Constructed"].size()
aggre_acquire
#Use size function to find how many "Year Constructed" has the same year as "Year Acquired"
#Refernce from :https://datascienceparichay.com/article/pandas-groupby-count-of-rows-in-each-group/
```

```python
aggre_acquire.index
#Year acquired
```

### the second visualization by using three different methods

## 2. total square footage as a function of building status ("Bldg Status")

```python
sum_footage=buildings.groupby("Bldg Status")["Square Footage"].sum()
```

```python
sum_footage
```

### first visualization

```python
type(sum_footage)
```

```python
sum_bar = sum_footage.plot.bar(x = 'Building Status', y ="Square Footage", rot = 0, logy = True, xlabel = 'Building Status', ylabel = 'Square Footage')
plt.suptitle('Total square footage as a function of building status ("Bldg Status")',fontsize=12, y = 1)
#Reference: https://pandas.pydata.org/docs/reference/api/pandas.Series.plot.bar.html
#Reference: https://kanoki.org/2019/09/16/dataframe-visualization-with-pandas-plot/ (for logy)
#Refernce: https://stackoverflow.com/questions/1388450/giving-graphs-a-subtitle-in-matplotlib (for suptitle)
```

### second visualization

```python
fig, ax = plt.subplots(figsize=(8,4))

ax.plot(sum_footage, marker='.')
ax.set_xlabel('Building Status')
ax.set_ylabel("Square Footage")
ax.set_yscale('log')
fig.suptitle('Total square footage as a function of building status ("Bldg Status")')
plt.show()
```

## write-up for question 2:
First, I calculate total 'Square Footage' of 'Bldg Status' by using groupby function. There are three building statuses: Abandon, In Progress, In Use. 

For the first visualization: I check the type of sum_footage is series. Thus, I use plot.bar function to plot a bar chart and set ylim for log.

For the second visualization: I set figure size first and plot the graph.
I observe that the volumn of square footage in abandon and in progress status are similar.

## 3. Median square footage in each Year Constructed as a function of Year Acquired

```python
median=buildings.groupby("Year Constructed")["Square Footage", "Year Acquired"].median()
```

```python
median
```

```python
fig = plt.gcf()
fig.set_size_inches(12, 8) #change seaborn plot size
sns.scatterplot(data=median, x='Year Acquired', y='Square Footage', hue='Year Constructed')
fig.suptitle('Median square footage in each Year Constructed as a function of Year Acquired')
#Reference: https://www.marsja.se/how-to-change-size-of-seaborn-plot/
```
